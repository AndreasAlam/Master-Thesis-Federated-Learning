```python
import matplotlib.pyplot as plt
```

### Functions 

```python
def loss_curve(list_vectors, legend_vec, epochs, txt):
    """ Plot loss against epochs.
    IN: list_vec (list), epochs (list int)
    OUT: - 
    """
    for vec in list_vectors:
        plt.plot(vec)
    
    #plt.ylim([-1, max(vec)])
    #plt.xlim([0, epochs/10])
    plt.ylabel('Average loss')
    plt.xlabel('Epoch')
    plt.legend(legend_vec, loc='upper right')
    plt.show()
    
    for index, l in enumerate(list_vectors):
        print('Min value', str(index+2), txt, ': ', str(min(l)))
```

```python
def aupcr(list_vectors, legend_vec, epochs, txt):
    """ Plot AUPCR against epochs.
    IN: list_vec (list), epochs (list int)
    OUT: - 
    """
    for vec in list_vectors:
        plt.plot(vec)

    plt.ylabel('Area under Profile Curve ratio (AUPCR)')
    plt.xlabel('Epoch')
    plt.legend(legend_vec, loc='upper left')
    plt.show()
    
    for index, l in enumerate(list_vectors):
        print('Max value', str(index+2), txt, ': ', str(max(l)))
```
